```python
import random
import string

def generate_random_password(length: int = 12, include_numbers: bool = False, include_symbols: bool = False) -> str:
    """
    generate a random a password.
    """
    characters: str = string.ascii_letters
    if include_numbers:
        characters += string.digits
    if include_symbols:
        characters += string.punctuation
        
    return "".join(random.choice(characters) for _ in range(length))

        
```

```python
generate_random_password(length=12,include_numbers= True, include_symbols= True)
```

```python
from typing import List, Optional

def generate_memorable_password(no_of_words: int = 10, separator: str = "-", capitalization: bool = False, vocabulary: Optional[List[str]] = None) -> str:
    """
    Generate a memorable password from a list of vocabulary words.
    """
    if vocabulary is None:
        vocabulary = ["Karo", "Leo", "Emad", "Amin", "Sanaz", "Mahdi", "Zeinab", "Mitra", "Shapoor", "Sina", "Janan"]
    password_words = random.sample(vocabulary, no_of_words)
    
    if capitalization:
        password_words = [word.capitalize() for word in password_words]
        
    return separator.join(password_words)
```

```python
generate_memorable_password(no_of_words=8, separator= "-", capitalization= True, vocabulary= None )
```

```python
def generate_pin(length: int = 6) -> str:
    """
    generate_pin a numeric pin code. 
    """
    return "".join(random.choice(string.digits) for _ in range(length))
```
